## centos-minimal
- centos minimal 默认不开起网络服务，需手动打开，步骤如下：
  ```
  #1.修改系统网络配置
  vi /etc/sysconfig/network-scripts/ifcfg-ens33

  #2.设置 ONBOOT=yes
  ONBOOT=yes

  #3.重启网络服务
  service network restart
  ```
- centos minimal 设置静态网络IP
  ```
  1.VMWare -> “编辑”菜单 -> “虚拟网络编辑器” -> “NAT设置” -> 查看IP、子网掩码、网关。
  如下图：
  ```

  <img width=80% height=80% src="imgs/01/001.png" alt="imgs/01/001.png" title="图1" />
  <img width=80% height=80% src="imgs/01/002.jpg" alt="imgs/01/002.jpg" title="图2" />

  ```
  2.修改系统网络配置
  vi /etc/sysconfig/network-scripts/ifcfg-ens33

  3.设置相关参数
  TYPE=Ethernet
  PROXY_METHOD=none
  BROWSER_ONLY=no
  #BOOTPROTO=dhcp
  BOOTPROTO=static
  DEFROUTE=yes
  IPV4_FAILURE_FATAL=no
  IPV6INIT=yes
  IPV6_AUTOCONF=yes
  IPV6_DEFROUTE=yes
  IPV6_FAILURE_FATAL=no
  IPV6_ADDR_GEN_MODE=stable-privacy
  NAME=ens33
  UUID=c1418054-f3ba-4dc2-9939-3fe54fcc5ca6
  DEVICE=ens33
  #ONBOOT=no
  ONBOOT=yes
  IPADDR=192.168.153.120
  NETMASK=255.255.255.0
  GATEWAY=192.168.153.2
  DNS1=114.114.114.114

  4.重启网络服务
  service network restart
  ```

### Centos7 服务基本设置
- hostname -> 
  ```
  # show hostname
  $ uname -n

  # set hostname
  $ hostnamectl set-hostname yourname
  ```
- SELinux ->
  ```
  # show status of selinux
  $ sestatus

  # disable selinux by editing the /etc/selinux/config file. As root, change the value of the SELINUX parameter in the config file as follows:
  SELINUX=disabled
  ```
- firewall ->
  ```
  # show firewall
  $ systemctl status firewalld

  # disable firewall
  $ systemctl stop firewalld.service
  $ systemctl disable firewalld.service
  ```

- 升级 ->
  ```
  # glibc升级到2.28
  $ wget https://ftp.gnu.org/gnu/glibc/glibc-2.28.tar.gz --no-check-certificate
  $ tar -xzvf glibc-2.28.tar.gz
  $ cd glibc-2.28
  $ mkdir build && cd build
  $ ../configure --prefix=/usr --disable-profile --enable-add-ons --with-headers=/usr/include --with-binutils=/usr/bin

  # 这一步时, 发生了错误, 提示大致为
  These critical programs are missing or too old: make compiler

  # 直接安装 GCC-8
  $ yum install -y devtoolset-8-gcc devtoolset-8-gcc-c++ devtoolset-8-binutils
  # 设置环境变量
  $ echo "source /opt/rh/devtoolset-8/enable" >> /etc/profile
  $ source /etc/profile

  $ wget https://ftp.gnu.org/gnu/make/make-4.3.tar.gz
  $ tar -xzvf make-4.3.tar.gz 
  $ cd make-4.3/
  # 安装到指定目录
  $ ./configure  --prefix=/usr/local/make
  $ make
  $ make install
  # 创建软链接
  $ cd /usr/bin/
  $ mv make make.bak # backup
  $ ln -sv /usr/local/make/bin/make /usr/bin/make

  # 进入之前的 `~/glibc-2.28/build`目录下
  $ cd /root/glibc-2.28/build
  $ ../configure --prefix=/usr --disable-profile --enable-add-ons --with-headers=/usr/include --with-binutils=/usr/bin
  $ make
  $ make install
  # 查询
  $ strings /lib64/libc.so.6 | grep GLIBC
  ...

  GLIBC_2.24
  GLIBC_2.25
  ```

## rpm

```
# 查看是否已安装
$ rpm -qa| grep ntp

# 安装已下载的rpm包，--nodeps 无视强制依赖
$ rpm -ivh --force --nodeps <rpm包名>
```

## yum

```
# 查看是否已安装
$ yum list installed | grep ntp

# 只下载 rpm 包，不执行安装，以下载 gdisk、net 工具包为例
$ yum --downloadonly --downloaddir=./ reinstall gdisk net-tools
```

## 主机操作

- 关机/重启
  ```
  # 立即重启
  $ shutdown --reboot now
  # 5分钟后重启
  $ shutdown --reboot +5

  # 立即关机
  $ shutdown --poweroff now
  # 5分钟后关机
  $ shutdown --poweroff +5
  ```
